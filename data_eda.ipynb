## Bivariate Analysis

```python
df1.head(2)
```

```python
df1.info()
```

```python
df1["Sales"].describe()
```

```python
bins = [0, 50, 200, df1["Sales"].max()]

labels = ["Low", "Medium", "High"]

df1["Sales_Category"] = pd.cut(
    df1["Sales"],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
# Analyze sales category and segment
sales_segment = (
    df1.groupby(["Segment", "Sales_Category"])
       .size()
       .reset_index(name="Count")
)


# plot bar chart
# sett figure
plt.figure(figsize=(8,6))

sales_segment.plot(
    kind="bar",
    figsize=(8,6),
    color=["skyblue", "orange", "green"],
    edgecolor="black"
)

plt.title("Sales Category by Segment")
plt.xlabel("Segment")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Sales Category")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.ticker as ticker
#grouping sales by segment
sales_segment = df1.groupby('Segment' )['Sales'].sum().reset_index()
# figure of bar chart
plt.figure(figsize = (7,6))
plt.bar(sales_segment['Segment'], sales_segment ["Sales"], color='skyblue', edgecolor='black')
plt.title('Total Sales By Segment')
plt.xlabel("Segment")
plt.ylabel('Total Sales')
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

```python
#grouping sales by region
sales_segment = df1.groupby('Region')['Sales'].sum().reset_index()
# figure of bar chart
plt.figure(figsize = (7,6))
plt.bar(sales_segment['Region'], sales_segment ["Sales"], color='skyblue', edgecolor='black')
plt.title('Total Sales By Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

## Analysis and Business Recommendation
From the analysis, we can see that the West region has the highest sales, followed by the East, Central, and South regions.

Business Recommendations
• A/B testing campaign should be using for underperforming regions to determine which ad campaign performs better in improving sales
• The marketing strategy used for the west region should be studied for underperforming regions
• Further analysis can be carried out to determine the exact reason for sales difference across regions
• Promo in form of discounts or free products should be given in regions like the south to see if it can improve sales

# Multivariate Analysis

```python
# sales by segment and shipping mode across regions
grouped = df1.groupby(['Region', 'Segment', 'Ship_Mode'])['Sales'].sum().reset_index()

grouped[ 'Segment_Ship_Mode'] = grouped[ 'Segment'] + '-' + grouped[ 'Ship_Mode' ]
pivot_df = grouped.pivot(index = 'Region', columns = 'Segment_Ship_Mode', values
='Sales')
pivot_df.plot(kind= 'bar', stacked=True, figsize=(6,8))
plt.title('Sales by Region and Ship_Mode')
plt.ylabel('Sales')
plt.xlabel( 'Region' )
plt.legend(titles= 'Segmemt - Ship Mode', bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt. show()
```

## Analysis and Business Recommendation

The west region leads all other regions, exceeding 700k driven mainly by consumers and corporations using the standard shipping class. The follows closely while the South lags behind significantly.
Across all regions, the standard class shipping mode dominates. This shows customers' preference for affordability over speed. The same day and first class contributes minimally, which also shows an untapped growth potential

Business Recommendation
• Premium shipping should be promoted through incentives and bundle offers
• Weak demands should be investigated and targeted promotions should be introduced
